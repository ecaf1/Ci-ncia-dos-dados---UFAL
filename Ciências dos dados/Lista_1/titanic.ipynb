```python
#Bibliotecas padrão para qualquer trabalho de anisé de dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carregar base principal pura
train = pd.read_csv("/home/edeilson/Área de Trabalho/Cursos-UFAL/Ciências dos dados/Lista_1/titanic/train.csv")
test = pd.read_csv("/home/edeilson/Área de Trabalho/Cursos-UFAL/Ciências dos dados/Lista_1/titanic/test.csv")

#Retirar colunas possivelmente inuteis 
train_clear = train.drop(columns=["Name","Cabin","Ticket"])
```

```python
#faremos uma analise do qunato uma duas variaveis podem se correlacionarem para determinar V ou M
V = train_clear[train_clear.Survived == 1]
M = train_clear[train_clear.Survived == 0]

plt.title(" Age vs sex")
plt.xlabel("Age")
plt.ylabel("Sex")
plt.scatter(M.Age, M.Sex, color = 'red', label = 'Morto', alpha= 0.3)
plt.scatter(V.Age, V.Sex, color = "lime", label = "Vivo", alpha = 0.3)
plt.legend()
plt.show()
```
